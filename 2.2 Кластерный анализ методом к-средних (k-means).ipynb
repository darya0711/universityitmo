```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv("Л•§•≠жл.dat", sep=";")
df.head()
```

```python
df.shape
```

```python
from sklearn.cluster import KMeans
# для кластеризации методом k-means необходимо задать k наперёд. Попробуем k = 4

model = KMeans(n_clusters=4, random_state=42, max_iter=300, n_init=50, verbose=0 ) # random_state - зерно датчика случайных чисел, n_init - количество раз разных положений центроидов
```

```python
model.fit(df) # подгонка модели по данным из df
```

```python
model.labels_ # результат кластеризации на данных df
```

```python
model.cluster_centers_ # координаты центров кластеров
```

```python
# Применим модель для разного числа кластеров, чтобы построить график каменистой осыпи

K = range(1, 11) # строим 10 моделей с числом кластеров от 1 до 10
models = [KMeans(n_clusters=k, random_state=42).fit(df) for k in K]
dist = [model.inertia_ for model in models] # inertia_ - возвращает сумму расстояний от каждой точки данных до центра ближайшего с ней кластера

plt.plot(K, dist, marker='o') # строим график каменистой осыпи
plt.xlabel('k (число кластеров)')
plt.ylabel('Сумма расстояний')
plt.title('Каменистая осыпь показывает оптимальное k')
plt.show()
```

```python
# На графике каменистой осыпи видно, что зависимость становится более пологой при разделении на 4 кластера
# Интерпретируем данные

model = KMeans(n_clusters=4, random_state=42, max_iter=300, n_init=10, verbose=0)
model.fit(df)
df['cluster'] = model.labels_ 
df.groupby('cluster').mean() # вычислим средние значения всех характеристик, сгруппировав данные по номеру кластера
```

```python
df.groupby('cluster').size() # подсчитаем объемы кластеров
```
